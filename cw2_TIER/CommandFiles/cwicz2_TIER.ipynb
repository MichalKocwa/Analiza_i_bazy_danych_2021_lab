```python
import pandas as pd
import numpy as np


s1 = 'data.csv'
```

```python
df = pd.read_csv(s1, header=None, delim_whitespace=True)
```

```python
df
#print(df.shape[0])
```

```python
# extracting date and type of measurement from 1 collumn 
id_l = []
rok_l = []
mies_l =[]
nazw_zmienn_l =[]
for i in range(df.shape[0]):
    id_l.append(df[0][i][0:11])
    rok_l.append(df[0][i][11:15])
    mies_l.append(df[0][i][15:17])
    nazw_zmienn_l.append(df[0][i][17:21])

    

series_1 = pd.Series(id_l)
series_2 = pd.Series(rok_l)
series_3 = pd.Series(mies_l)
series_4 = pd.Series(nazw_zmienn_l)

df.drop(0, axis='columns', inplace=True)

#  creating and inserting new collumns containing newly obtained informations 

df.insert(0, 'Id', series_1)
df.insert(1, 'rok', series_2)
df.insert(2, 'miesiac', series_3)
df.insert(3, 'element', series_4)

df
```

```python
ser_vb = df['rok']+"-"+df['miesiac']
df.drop("rok", axis='columns', inplace=True)
df.drop("miesiac", axis='columns', inplace=True)
df.insert(1, 'data', ser_vb)

df2 = df.melt(id_vars = ["Id","data","element"])
```

```python
df2['variable'] = df2['variable'].astype(str)
df2ser_vb = df2['data']+"-"+df2['variable']
df2.drop("variable", axis='columns', inplace=True)
df2.drop("data", axis='columns', inplace=True)
df2.insert(1, 'data', df2ser_vb)
df2.sort_index(axis=0,ascending=True)
df2 = df2.sort_values(by="data")

df2.reset_index(inplace = True, drop = True)

df2.head(15)
```

```python
df2.to_csv('data_tidy.csv',index=True)
```
