```python
import pandas as pd
import numpy as np

s1 = 'C:\\Projects\\STUDIA\\ABD\\Lab1\\data.csv'
#s2 = 'C:\\Projects\\STUDIA\\ABD\\Lab1\\data2.txt'
```

```python
df = pd.read_csv(s1, header=None, delim_whitespace=True)
```

```python
df
#print(df.shape[0])
```

```python
id_l = []
rok_l = []
mies_l =[]
nazw_zmienn_l =[]
for i in range(df.shape[0]):
    id_l.append(df[0][i][0:11])
    rok_l.append(df[0][i][11:15])
    mies_l.append(df[0][i][15:17])
    nazw_zmienn_l.append(df[0][i][17:21])
```

```python
series_1 = pd.Series(id_l)
series_2 = pd.Series(rok_l)
series_3 = pd.Series(mies_l)
series_4 = pd.Series(nazw_zmienn_l)

df.drop(0, axis='columns', inplace=True)

df.insert(0, 'Id', series_1)
df.insert(1, 'rok', series_2)
df.insert(2, 'miesiac', series_3)
df.insert(3, 'nazw. zmienn.', series_4)

df_kon = df.rename(columns = {1: 'd1', 2: 'd2',3:"d3", 4: 'd4', 5: 'd5',6:"d6", 7: 'd7', 8: 'd8',9:"d9", 10: 'd10', 11: 'd11',12:"d12",
                                13: 'd13', 14: 'd14',15:"d15",16: 'd16', 17: 'd17',18:"d18",19: 'd19', 20: 'd20',21:"d21",22: 'd22', 23: 'd23',24:"d24",
                                25: 'd25', 26: 'd26',27:"d27",28: 'd28', 29: 'd29',30:"d30",31: 'd31'}, inplace = False)
df_kon
```
