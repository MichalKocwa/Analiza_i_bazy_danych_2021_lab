```python
#Micha≈Ç Kocwa

##Zadanie 3

import numpy as np
import matplotlib.pyplot as plt


def fun1(x):
  return x*x + 5


x = np.arange(-1,1,0.01)

plt.plot(x,fun1(x), label = "x^2+5")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('wykres 3.1')
plt.legend()
plt.show()
```

```python
x = np.arange(-6,6,0.01)
plt.plot(x,fun1(x), label = "x^2+5")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('wykres 3.2')
plt.legend()
plt.show()
```

```python
x = np.arange(0,5,0.01)
plt.plot(x,fun1(x), label = "x^2+5")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('wykres 3.3')
plt.legend()
plt.show()
```
